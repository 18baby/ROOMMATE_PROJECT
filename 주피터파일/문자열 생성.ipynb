```python
partner_df = df.iloc[:, 19:]
partner_df.head()
partner_df
```

```python
partner_df = partner_df.drop('설문지 응답하신분들중에 추첨으로 상품을 전달 드릴 예정입니다!!\n원하시는 분들은 아래 연락처를 남겨주시면 당첨 시 연락 드리겠습니다!!', axis=1)
partner_df.head()
```

```python
start = ["나이대는 ", "같은 단과대 ", "성격유형은 ", "주말거주는 ", "평일 거주는 ", 
         "", "음주는 ", "식사는 ", "배달 음식은 ", "기상시간은", "취침시간은 ", "취침 예민도는 ", 
         "잠버릇은 ", "청소 빈도는 ", "샤워시간대는 ", "추가정보는 "]

print(len(start))
print(len(partner_df.columns))
```

```python
for index, row in partner_df.iterrows():
    #print(f"인덱스: {index}")
    text = ""
    for i in range(len(row)):
        #print(start[i])
        #print(row[i])
        if row[i] == "상관 없음":
            continue
        if not isinstance(row[i], str):
            break
        text = text + start[i] + row[i] +", "
    print(text)
    partner_df.loc[index, 'text'] = text
```
