```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
```

```python
main_df = pd.read_csv('설문지_응답.csv')
df = main_df.iloc[:, 1:34]
df.shape
```

```python
df.columns
```

```python
# 변경할 열이름
col_list = ['email', 'sex', 'age', 'dorm_select', 'college_of', 'personality', 'weekend_stay', 
            'weekday_stay', 'smoke', 'alchol', 'm_how_eat', 'how_eat_in', 'wake_up', 'm_sleep', 
            'sleep_sensitive', 'sleep_habit', 'clean_period', 'shower_timezone', 'w_age_range', 
            'w_diff_college_of', 'w_personality', 'w_weekend_stay', 'w_weekday_stay', 'w_smoke', 
            'w_alchol', 'w_how_eat', 'w_how_eat_in', 'w_wake_up', 'w_sleep', 'w_sleep_sensitive', 
            'w_sleep_habit', 'w_clean_period', 'w_shower_timezone']

df.columns = col_list
df.head()
```

```python
# 나이 전처리
df.loc[df['age'] < 20, 'age'] = 20
df.loc[df['age'] > 28, 'age'] = 28

# 잘 들어갔는지 확인
df['age'].unique()
```

```python
# primary key인 이메일 삭제
col_list.remove('email')
print(col_list)
print(len(col_list))

# 범주형 변수만 있는 DF -> 사실상 전체 DF
df_cat = df[col_list].copy()         # 문자열로 저장된 DF
df_label = df[col_list].copy()       # 라벨 인코딩 된 DF
```

```python
df_label['age'] = df_label['age']%10
df_cat['age'] = df_cat['age']%10
```

```python
# 상관없음 전처리
replace_dict = {'상관 없음': '0'}
df_label.replace(replace_dict, inplace=True)
```

```python
df_label
```

```python
# LabelEncoder를 객체로 생성합니다.
le = LabelEncoder()

for col in df_label.columns:
    if df_label[col].dtype == 'object':  # 오브젝트 타입의 열만 인코딩
        
        if col == 'wake_up':
            df_label.loc[df_label[col]=='6시~8시', col] = '1'
            df_label.loc[df_label[col]=='8시~10시', col] = '2'
            df_label.loc[df_label[col]=='10시~12시', col] = '3'
            df_label.loc[df_label[col]=='12시 이후', col] = '4'
        elif col == 'w_wake_up':
            df_label.loc[df_label[col]=='6시~8시', col] = '1'
            df_label.loc[df_label[col]=='8시~10시', col] = '2'
            df_label.loc[df_label[col]=='10시~12시', col] = '3'
            df_label.loc[df_label[col]=='12시 이후', col] = '4'
        
        df_label[col] = le.fit_transform(df_label[col])

df_label   # 라벨 인코딩 결과
```

```python
plus = ['sex', 'dorm_select', 'personality', 'weekend_stay', 'weekday_stay', 'smoke', 'alchol', 'm_how_eat', 'how_eat_in', 'wake_up', 'm_sleep', 'sleep_sensitive', 'sleep_habit', 'clean_period', 'shower_timezone']
print(len(plus))

df_label[plus] += 1   # want_ 정보랑 맞추기 위해서 1 더하기
```

```python
df_label
```

```python
label_df = pd.concat([df['email'], df_label], axis=1)
label_df
```

```python
label_df.to_csv('main_data.csv')
```
